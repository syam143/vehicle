```python
weather = pd.read_csv('vehicle.csv')
```

```python
weather.info()
```

```python
sns.barplot(weather['fm'],weather['Mileage'])
```

```python
sns.displot(weather['fm'])
```

```python

sns.distplot(weather['fm'], kde=False, rug=True);
```
